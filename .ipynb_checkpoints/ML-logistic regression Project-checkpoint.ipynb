# Logistic Regression Project

In this supervised machine learning project, I have used a fake advertisement dataset to train a model. the goal of the model is to predict if a user will click on a 'ad' or not based on the user features. and because it is a binary classification problem, I have used Logistic Regression model to fit the data.


### About the data

* 'Daily Time Spent on Site': consumer time on site in minutes
* 'Age': cutomer age in years
* 'Area Income': Avg. Income of geographical area of consumer
* 'Daily Internet Usage': Avg. minutes a day consumer is on the internet
* 'Ad Topic Line': Headline of the advertisement
* 'City': City of consumer
* 'Male': Whether or not consumer was male
* 'Country': Country of consumer
* 'Timestamp': Time at which consumer clicked on Ad or closed window
* 'Clicked on Ad': 0 or 1 indicated clicking on Ad


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
ad_df=pd.read_csv('advertising.csv')
```

```python
ad_df.head()
```

### Checking the data for missing values

I have used heatmap to check if the dataset contains any missing data or not and it seems like no data is missing

```python
sns.heatmap(ad_df.isnull(),cmap='viridis')
```

### visualizing the data

I have used pair plot to see the relationship between the columns

```python
sns.set_style('whitegrid')
sns.set_palette('Accent')
sns.pairplot(ad_df,hue='Clicked on Ad')
```

### Adding new features

'Ad Topic Line' column shows the Headline of the advertisement. the headline size can be an important factor for a users decision making. for example, if a headline is too short then the user may not be able to clearly understand what is the ad is about hence, not click the ad. on the other hand if the heading is too broad the user may not bother to read the full headline and ignore the ad. so I have added a new feature column which will hold the total number of word that is in the corresponding Ad Topic Line column.  

```python
ad_df['Topic size'] = ad_df['Ad Topic Line'].apply(lambda x: len(x))
```

```python
ad_df.head()
```

from the plot bellow the new 'Topic size' column data seems like normally distributed. and for liner models like logistic regression they tends to perform better with normally distributed data.


```python
plt.figure(figsize=(10,5))
sns.distplot(ad_df['Topic size'])
```

```python
ad_df.drop('Ad Topic Line',axis=1,inplace=True) # dropping the 'Ad Topic Line' column
```

```python
ad_df['Country'].value_counts()
```

dropping 'Country' and 'City' columns because of their large variation in data. They are note giving any useful insights about the data

```python
ad_df.drop(['Country','City'],axis=1,inplace=True) 
```

```python
ad_df
```

converting the 'Timestamp' column in to three new columns converting the date time string to a numerical value

```python
ad_df['Timestamp'] = pd.to_datetime(ad_df['Timestamp'])
```

```python
ad_df['Hour'] = ad_df['Timestamp'].apply(lambda time: time.hour)
ad_df['Month'] = ad_df['Timestamp'].apply(lambda time: time.month)
ad_df['Day of Week'] = ad_df['Timestamp'].apply(lambda time: time.dayofweek)
```

```python
ad_df.head()
```

```python
ad_df.drop(['Timestamp'],axis=1,inplace=True) # dropping the 'Timestamp' column
```

```python
ad_df.head()
```

```python
for column in ad_df.columns:           # checking the distribution of each colum
    plt.figure()             
    sns.distplot(ad_df[column])
```

### Logistic Regression
Now it's time to do a train test split, and train the model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(ad_df.drop(['Clicked on Ad'],axis=1),
                                                    ad_df['Clicked on Ad'], 
                                                    test_size=0.30, random_state=101)
```

fitting the train data into the model

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test) # predicting from the test data
```

```python
from sklearn.metrics import classification_report
```

### Classification report

```python
print(classification_report(y_test,predictions))
```
